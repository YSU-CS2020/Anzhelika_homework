Question 6. Write code that checks the data type of a variable and saves the result as a string in `value_type`, using match/case statements:

"Integer", "String", or "Boolean"

```python
value = """
Մի մրահոն աղջիկ տեսա
Ռիալտոյի կամուրջին,
Հորդ մազերը – գետ գիշերվա,
Եվ հակինթներ՝ ականջին:
"""

# Your code here
match value:
    case int() if not isinstance(value, bool):
        value_type = "Integer"
    case str():
        value_type = "String"
    case bool():
        value_type = "Boolean"
    case _:
        value_type = "Other"

# ------------------------

assert value_type == "String"
print("Question 6 passed!")
```
